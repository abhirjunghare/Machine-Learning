```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import os
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']], columns = np.append(cancer_dataset['feature_names'],['target_names']))
cancer_df                                                                  
```

```python
x = cancer_df.iloc[:,0:-1]
x
```

```python
y = cancer_df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 50)
x_test
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train,y_train)
```

```python
classifier.score(x_test,y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier_m = MultinomialNB()
classifier_m.fit(x_train,y_train)
```

```python
classifier_m.score(x_test,y_test)
```

```python
from sklearn.naive_bayes import BernoulliNB
classifier_b =  BernoulliNB()
classifier_b.fit(x_train, y_train)
```

```python
classifier_b.score(x_test,y_test)
```
